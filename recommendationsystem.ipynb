# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

# Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Title']>200
imp_users = x[x].index
```

```python
len(imp_users)
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(imp_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
len(famous_books)
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index ='Book-Title',columns = 'User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(pt)
```

```python
similarity.shape
```

```python
similarity[0].shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar = sorted(list(enumerate(similarity[index])),key = lambda x:x[1],reverse = True)[1:6]
    
    for i in similar:
        print(pt.index[i[0]])
```

```python
sorted(list(enumerate(similarity[0])),key = lambda x:x[1],reverse = True)[1:6]
```

```python
recommend('Zoya')
```

```python
recommend('A Bend in the Road')
```
